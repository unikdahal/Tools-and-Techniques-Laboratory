```python
sentence = input("Enter a sentence: ")
new_sentence = ""
for char in sentence:
    if char.isupper():
      new_sentence += char.lower()
    elif char.islower():
      new_sentence += char.upper()
    else:
      new_sentence += char

print(new_sentence)
```

```python
m = int(input("Enter the number of rows: "))
n = int(input("Enter the number of columns: "))
array = [[0 for _ in range(n)] for _ in range(m)]

for i in range(m):
    for j in range(n):
        array[i][j] = i * j

for i in range(m):
    for j in range(n):
        print(array[i][j], end=" ")
    print()
```

```python
string = input("Enter a string: ")
digits = 0
letters = 0
for char in string:
    if char.isdigit():
        digits += 1
    elif char.isalpha():
        letters += 1
print("Digits:", digits)
print("Letters:", letters)
```

```python

lower = 900
upper = 1000

for num in range(lower, upper + 1):
  if num > 1:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
      if (num % i) == 0:
        is_prime = False
        break
    if not is_prime:
      print(num)

```
